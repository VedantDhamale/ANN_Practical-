Write a Python program to plot a few activation functions that are being used in neural network

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.linspace(-10,10,100)
```

Linear activation function :

It simply computes a weighted sum of the input values (plus a bias), which means the output is directly proportional to the input.

Range:(-∞, +∞).



```python
#linear
plt.title("Linear Activation function")
plt.plot(x,x)
```

sigmoid activation function:

non-linear activation function, popular in binary classification tasks 

range is (0, 1).



```python
#sigmoid
plt.title("Sigmoid Activation function")
plt.plot(x,1/(1+np.exp(-x)))
```

hyperbolic tangent (tanh) activation function:

non-linear activation function 

range is (-1, 1).

```python
#tanh
plt.title("Tanh Activation Function")

plt.plot(x,np.tanh(x))
```

Binary Activation Function:

produce binary outputs, typically 0 or 1, based on a threshold.

0, if x<threshold
1, if x≥threshold
​

```python
plt.plot(x,np.heaviside(x,1))
```

ReLU Activation Function:

by outputting the input directly if it's positive, and zero otherwise

non-linear activation function

range of ReLU is [0,+∞]



```python

plt.plot(x,np.maximum(0,x))
```

Softmax activation function:

especially in multi-class classification tasks where it transforms raw scores into meaningful probabilities that can be used for decision-making.

non-linear activation function

range of (0, 1)

```python
#softmax
plt.title("SoftMax Function")
plt.plot(x,np.exp(x)/np.sum(np.exp(x)))
```

Swish activation function:

non-linear activation function

like ReLU (Rectified Linear Unit)

The range of the Swish activation function, like many other activation functions, depends on its definition.

```python
plt.title("Swish Activation function")
plt.plot(x,x/(1+np.exp(-x)))
```

Softsign Activation Function: non-linear function 

It is similar to the sigmoid function in terms of its properties but has some differences in behavior. 

range between -1 and 1

```python
plt.plot(x,x/(1+abs(x)))
plt.title("Softsign Activation Function")
```

Tanh Strink function

```python
plt.plot(x,x-np.tanh(x))
```

Leaky ReLU Function

```python
plt.plot(x,np.maximum(0.1*x,x))
```

soft plus

```python
plt.plot(x,np.log(1+np.exp(x)))
```
