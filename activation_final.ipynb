```python
import numpy as np
```

```python
x=np.linspace(-10,10,100)
```

```python
from matplotlib import pyplot as plt
```

```python
#linear
plt.plot(x,x)
```

```python
#binary
plt.plot(np.heaviside(x,1))
```

```python
#sigmoid
plt.plot(x,1/(1+np.exp(-x)))
```

```python
#swish
plt.plot(x,x/(1+np.exp(-x)))
```

```python
#soft sign
plt.plot(x,x/(1+abs(x)))
```

```python
#tanh
plt.plot(x,np.tanh(x))
```

```python
#tanshirnk
plt.plot(x,x-np.tanh(x))
```

```python
#relu
plt.plot(x,np.maximum(0,x))
```

```python
#leaky relu
plt.plot(x,np.maximum(0.1*x,x))
```

```python
#soft max
plt.plot(x,np.exp(x)/np.sum(np.exp(x)))
```
